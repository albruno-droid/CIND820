## Retrieve Data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('/content/BeijingPM20100101_20151231.csv')
df.head()
```

## Initial Summary inspection

Illustrate the first few rows of the DataFrame and check the data types and percentage of missing values. Similarly, generate and display descriptive statistics for numerical columns to better understand their distribution and summary measures.


```python
display(df.info())
display(df.isnull().sum()/ df.shape[0] * 100)
```

```python
display(df.describe())
```

## Visualize data distribution




```python
# Frequency distribution
plt.figure(figsize=(15, 6))
plt.hist(df['PM_US Post'], bins=20, color='skyblue', edgecolor='black')
plt.title('Frequency Distribution of PM2.5')
plt.xlabel('PM2.5 concentration')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Visualize Outlier Boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['PM_US Post'])
plt.title('Box Plot of PM_US Post')
plt.xlabel('PM2.5 concentration')
plt.grid(True)
plt.show()
```

## Correlation analysis
Examine correlations between relevant columns. Visualize the correlation using a heatmap


```python
# Select relevant numerical columns for correlation analysis
relevant_cols = ['PM_US Post', 'PM_Dongsi', 'PM_Dongsihuan', 'PM_Nongzhanguan', 'DEWP', 'HUMI', 'PRES', 'TEMP', 'Iws', 'precipitation', 'Iprec']
correlation_matrix = df[relevant_cols].corr()

# Display the correlation matrix using a heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix of Relevant Numerical Columns')
plt.show()
```

## Summary:

### Data Analysis Key Findings

*   The dataset contains 52,584 entries and 18 columns, including PM measurements from four locations and various meteorological features.
*   Several columns, particularly the PM measurements (`PM_Dongsi`, `PM_Dongsihuan`, `PM_Nongzhanguan`, `PM_US Post`), have a significant number of missing values.
*   The distribution of 'PM\_US Post' is right-skewed with a long tail, indicating a higher frequency of lower PM values and fewer occurrences of very high values, as shown by the histogram and box plot. The box plot also highlights potential outliers in 'PM\_US Post'.
*   Strong positive correlations were observed among the different PM measurement columns (e.g., the correlation between 'PM\_US Post' and 'PM\_Dongsi' is approximately 0.94).
*   There are notable correlations between PM measurements and meteorological factors: 'PM\_US Post' shows a positive correlation with 'DEWP' (approximately 0.68) and 'HUMI' (approximately 0.19), and negative correlations with 'PRES' (approximately -0.11) and 'TEMP' (approximately -0.24).
